- “어떤 부분을 변화시키면 우리의 핵심 지표(ex. 매출)가 올라갈까?”

## 📈회귀선

- 회귀선은 두 변수 사이의 관계를 하나의 선으로 나타냄
- $y = ax + b$
    - x: 독립변수
    - a: 기울기(회귀계수)
        - 독립변수가 1증가할 때 종속변수가 얼마나 증가하는지 나타내는 값
    - b: 절편
        - 독립변수가 0일 때, 종속변수의 값

- 독립변수(x): 고객이 특정 상품 페이지를 본 횟수
- 종속변수(y): 고객의 해당 상품 구매 금액
- 기울기(a): 고객이 상품 페이지를 한 번 더 방문할 때 평균적으로 구매 금액이 얼마나 증가하는지 나타냄.
    - 예: a=20,000이라면 고객이 상품 페이지를 한 번 더 보면 평균적으로 구매 금액이 20,000원이 증가한다는 의미입니다.
- 절편(b): 고객이 상품 페이지를 전혀 보지 않아도 발생할 수 있는 구매 금액
    - 예: 광고, 추천 등으로 인해 발생한 구매 금액

- 회귀선은 "상품 페이지를 많이 본 고객일수록 해당 상품을 더 많이 구매한다"는 경향성을 설명합니다.

$$
y=20,000x+50,000
$$

- 고객이 상품 페이지를 0번 본 경우, 예상 구매 금액은 50,000원(b)입니다.
- 고객이 상품 페이지를 5번 방문한 경우

$$
y=20,000*5+50,000 = 150,000
$$

- 페이지를 5번 방문했을 때 예상 구매 금액은 150,000만원

- 최소제곱법(Ordinary Least Squares, OLS)
- 회귀계수와 절편을 구하면,

## 회귀에서의 가설검정

1. 회귀식의 표현
- 독립변수(x)와 종속변수(y) 사이의 관계를 나타내는 회귀계수를 확인하는 것이 목표
- 🔎 회귀계수(${\beta_i}$, Coefficients) 값을 찾아라!

$$ y = \beta_0 + \beta_1 x $$

2. 가설설정
- 귀무가설(Null Hypothesis)
    - 독립변수가 종속변수에 영향을 미치지 않는다.
    - 즉, 회귀계수는 0이다.
- 대립가설(Alternative Hypothesis)
    - 독립변수가 종속변수에 영향을 미친다.
    - 즉, 해당 회귀계수는 0이 아니다.

---

3. **최소자승법(OLS, Ordinary Least Squares)**

- 회귀분석에서 최적의 회귀선을 찾기 위해 사용하는 방법
- 잔차(Residual)의 제곱합을 최소화하는 회귀계수를 구하는 것이 목표

1. 잔차(Residual)
- 실제값과 예측된 값 사이의 차이
- ${e_i = y_i - \hat{y_i}}$

2. 잔차제곱합(Sum of Squared Residuals, SSR) ⭐️⭐️⭐️
    - SSR
    - RSS
- 잔차의 제곱을 모두 더한 값
- 잔차의 제곱합을 최소화하는 회귀계수를 구하는 것이 목표
- 실제값과 예측값 사이의 차이가 전반적으로 얼마나 큰지를 의미하는 지표
- 계산 결과로 각 독립변수에 대한 회귀계수 값이 구해집니다.

- 예를 들어, 시험점수에 대한 회귀분석을 진행했다고 가정하겠습니다.

| 이름   | 통계 점수 | 코딩 점수 | 종합평가(실제점수) | 종합평가(예측점수) | 오차 (실제-예측) |
|--------|-----------|-----------|--------------------|--------------------|------------------|
| 두란   | 60        | 80        | 70                 | 75                 | -5               |
| 파이카 | 90        | 95        | 90                 | 95                 | -5               |
| 우너   | 70        | 80        | 85                 | 80                 | 5                |
| **합계** | **220**   | **255**   | **245**            | **250**            | **-5**           |



$$
SSR = \sum_{i=1}^{n} e_i^2 = \sum_{i=1}^n (y_i-\hat{y_i})^2
$$


3. 잔차의 분산(Residual Variance)
- 잔차제곱합을 자유도로 나눈 값
$$
s^2 = \frac{SSR}{n - p}
$$
- $n$: 데이터의 수
- $p$: 추정하려고 하는 계수(parameter)의 갯수


4. 표준오차(SE, Standard Error)

$$
SE(\hat{\beta_1}) = \sqrt{\frac{s^2}{\sum_{i=1}^{n}(x_i - \bar{x})^2}}
$$

- 회귀계수의 추정치의 신뢰성을 평가하는 데 사용
- 표준오차가 작을수록 해당 회귀계수의 추정이 모집단의 실제 값을 더 정확히 반영할 가능성이 높다는 것을 의미합니다.
- 데이터가 덜 분산되어 있으며, 독립변수와 종속변수 간의 관계가 더 명확하다는 의미입니다.

---

4. t-통계량(t-stat) 계산
- 최소자승법(OLS)을 활용하여 독립변수의 회귀계수($\hat{\beta_i}$) 값을 찾았습니다.
- 찾아낸 회귀계수에 대한 t-통계량을 계산합니다.
    - 회기계수에 대한 t-통계량은 회귀계수가 귀무가설 하에서 얼마나 유의미한지 검정하는 데 사용됩니다.
- 여기서 얻어낸 t-통계량이 t분포 상에서 어느 정도 위치에 있는지 확인합니다.
    - 회귀계수가 0이라는 가정하에 얼마나 멀리 떨어져 있는지를 확인합니다.

- t-통계량 공식
$$
t = \frac{coef}{std err}
$$

- 회귀계수가 0 즉, 귀무가설을 가정할 때, 해당 회귀계수가 실제로 0과 얼마나 차이가 나는지를 나타내는 지표
- t-값이 크면 회귀계수가 0일 가능성이 낮아지므로, 해당 변수는 유의미하다고 해석합니다.

$$
t = \frac{\hat{\beta_i} - \beta_0}{SE(\hat{\beta_i})}
$$

- 추정된 회기계수: $\hat{\beta}$
- 귀무가설에서 가정한 회귀계수: ${\beta_0}$ = 0
    - 즉, 식에서 생략가능


5. p값 계산($P > |t|$) ⭐️⭐️⭐️
- t-통계량을 기반으로 p값을 계산합니다.
- 계산된 p값은 t-분포를 활용하여 구하며, 유의수준(일반적으로α=0.05)과 비교합니다.
- p값 계산 공식

$$
p=2*P(T > |t|)
$$

- 여기서 $P(T > |t|)$는 t-분포의 누적분포함수(CDF)를 사용해 계산됩니다.



6. 가설 기각 여부 판단
- p값이 유의수준보다 작으면 귀무가설 기각, 대립가설 채택
- p값이 유의수준보다 크면 대립가설 기각, 귀무가설 채택

7. 95% 신뢰구간

|Lower 95%|Upper 95%|
|---------|---------|
|402|810|

- 회귀계수가 이 범위 안에 있을 확률이 95%임을 의미합니다.

8. 결과 해석 ⭐️⭐️⭐️
- p값이 작아 귀무가설을 기각한 경우, 해당 독립변수는 종속변수에 영향을 미치는 중요한 요인으로 판단됩니다.
- p값이 유의수준보다 큰 경우, 해당 독립변수는 종속변수에 영향을 미치지 않거나, 데이터에서 그 영향을 확인할 수 없다고 판단합니다.

```
Q: 선형회귀분석에 대해 설명해보세요.

A: 🙋🏻‍♂️
회귀분석에서는 먼저 최소자승법(OLS)을 활용해 회귀계수(β) 값을 추정합니다.
그런 다음, 추정된 회귀계수에 대해 t-통계량을 계산하여 귀무가설(예: 해당 계수가 0이다)을 검정합니다.
마지막으로, 계산된 t-통계량을 기반으로 t-분포에서 p-값을 구하고, 이를 유의수준과 비교하여 회귀계수가 통계적으로 유의미한지 판단합니다.
```
