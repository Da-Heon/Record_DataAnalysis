판다스의 경우 컬럼에 일괄적으로 데이터 가공을 하는 것이 속도 면에서 더 빠르나 복잡한 데이터 가공이 필요할 경우 어쩔 수 없이 apply lamnda를 이용합니다.

- 입력값의 제곱 값을 구해서 반환하는 함수를 만들어보겠습니다

````python
def get_square(a):
    return a ** 2

get_square(3)
````

- 위의 함수는 함수명과 입력 인자를 먼저 선언하고 이후 함수 내에서 입력 인자를 가공한 뒤 결괏값을 return과 같은 문법으로 반환해야 합니다
- lambda는 이러한 함수의 선언과 함수 내의 처리를 한 줄의 식으로 쉽게 변환하는 식입니다.
- 위 함수를 변환해보겠습니다

````python
square = lambda x : x ** 2

print(f'10의 제곱은 {square(10)}입니다.')
````

- : 의 왼쪽에 있는 x는 입력 인자를 가리키며, 오른쪽은 입력 인자으 계산식입니다. 오른쪽의 계산식은 결국 반환 값을 의미합니다.
- 여러개의 입력 인자를 사용해야 할 경우는 map() 함수를 이용합니다.

````python
a = [1, 2 ,3]

abc = map(lambda x : x ** 2, a)
print(f' 1, 2, 3의 각 제곱근은 {list(abc)}입니다')
````

````python
df = pd.read_csv('/content/drive/MyDrive/데이터분석 연습/pizzas.csv')
df.head()
````

````python
# 피자네임 컬럼값을 인자로 설정하고 그 값의 이름의 갯수를 나타내는 새로운 컬럼 생성
df['name_len'] = df['pizza_name'].apply(lambda x: len(x))
df.head()
````

- if else문을 사용해보자

- size가 M 또는 L이면 BIG, 나머지는 SMALL을 나타내는 새로운 컬럼 생성

````python
df['size_detail'] = df['size'].apply(lambda x: 'Big' if x == 'M' or x == 'L' else 'Small')
df.head()
````

- 가격이 5000원 이하면 low, 10000원에서 20000 사이이면 soso 20000 이상이면 expensive로 분류하는 컬럼을 만들어 보겠습니다

````python
df['price_detail'] = df['price'].apply(lambda x : 'low' if x <= 5000 else ('soso' if x <= 20000 else 'expensive'))
df.head()
````
